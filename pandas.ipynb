## Adding Columns


### SONG COLUMNS

```python
def get_season(month):
    if month in [12,1,2]:
        return 'winter'
    elif month in [3,4,5]:
        return 'spring'
    elif month in [6,7,8]:
        return 'summer'
    elif month in [9,10,11]:
        return 'fall'
    else: return np.nan

def is_christmas_song(name):
    christmas_words =  ['navidad', 'carol','christmas', 'snow', 'santa', 'cold', 'jingle','jolly', 'mistletoe', 'snowman', 'sleigh', 'deck the halls', 'candy cane', 'xmans', 'silent night', 'holiday', 'holidays', 'rudolph', 'reindeer', 'noel', 'grinch', 'drummer boy']
    name= name.lower()
    return any(word in name for word in christmas_words)


def popularity_buckets(popularity):
    if popularity >=80:
        return 'viral'
    elif popularity >= 60:
        return 'popular'
    elif popularity>= 30:
        return 'moderate'
    elif popularity>=0:
        return 'niche'
    else: return np.nan
```

```python

#Engagement Efficiency
songs['engagement_efficiency']= songs.apply(lambda x:round(x['song_listeners']/ (x['popularity']),2) if x['popularity']>0 else 0, axis=1)

#VITALITY SCORE
songs['song_virality'] = songs.apply(lambda x: round((x['engagement_ratio']* x['popularity'])/100,2) if x['engagement_ratio'] is not None else 0, axis=1)

#NORMALIZED SONG LISTENERS
listeners = songs['song_listeners']
songs['normalized_song_listeners']= (listeners- listeners.min())/(listeners.max()- listeners.min())


#DATE COLUMNS
release_date = songs['release_date']
songs['release_year']= pd.to_datetime(release_date, errors='coerce').dt.year.fillna(release_date.str[:4].astype(int, errors='ignore').fillna(9999))
songs[songs['release_year']!=9999]
songs['release_month']= pd.to_datetime(release_date, errors='coerce').dt.month
songs['release_decade']= (songs['release_year']//10)*10
songs['release_season']= songs['release_month'].apply(get_season)
songs['is_christmas_song'] = songs.apply(lambda row: is_christmas_song(row['song_name']) and row['release_month'] in [11, 12], axis=1)
songs['song_age_years'] = datetime.now().year - songs['release_year']


#Categorical Variables:
songs['popularity_bins']= songs['popularity'].apply(popularity_buckets)

zscore_cols=['song_listeners', 'song_virality', 'popularity', 'song_age_years', 'duration_minutes', 'engagement_ratio']
for col in zscore_cols:
    data= songs[col]
    songs[f'{col}_zscore']= (data-data.mean())/data.std()
    

display(songs.sample(10))



```

### AUDIO FEATURE COLUMNS

```python
audio_features['audio_complexity']= (audio_features['spectral_centroid']/1000)+audio_features['zero_crossing_rate']
audio_features['harmonic_balance'] = (audio_features['harmonic_ratio']- audio_features['percussive_ratio'])
audio_features['dance_energy_index']= (audio_features['danceability'] * audio_features['energy']* audio_features['bpm'])/200

zscore_cols=['danceability', 'energy', 'spectral_centroid', 'zero_crossing_rate', 'bpm']
for col in zscore_cols:
    std= audio_features[col].std()
    mean= audio_features[col].mean()
    audio_features[col+"_zscore"]= (audio_features[col]-mean)/std


display(audio_features.sample(10))
```

### ARTIST COLUMNS


```python


#ARTIST AVG POPULARITY
artist_agg= songs.groupby('artist_id').agg({'popularity': ['mean', 'std', 'max', 'min'], 
'song_age_years': 'mean', 
'song_listeners': 'mean',
'engagement_ratio': 'mean',
'duration_minutes': 'mean',
'song_id': 'count',
'release_year': ['min', 'max', 'mean']
}).reset_index()
artist_agg.columns = ['artist_id', 'avg_popularity', 'popularity_std', 'max_popularity', 'min_popularity', 
'avg_song_age', 'avg_artist_listeners', 'avg_engagement_ratio', 'avg_duration_minutes', 'song_count', 'first_release', 'latest_release', 'avg_release_year']
print(artist_agg.describe())
artists=artists.merge(artist_agg, how='left', on= 'artist_id')
display(artists)
```

### ALBUM COLUMNS


```python
album_pop= songs.groupby('album_id').agg({'popularity': ['mean', 'std', 'max', 'min'], 'song_age_years': 'first'}).reset_index()
album_pop.columns = ['album_id', 'avg_popularity', 'popularity_std', 'max_popularity', 'min_popularity','album_release_year']
print(album_pop.describe())
albums=pd.merge(albums, album_pop, how='right', on= 'album_id')
display(albums)
```
