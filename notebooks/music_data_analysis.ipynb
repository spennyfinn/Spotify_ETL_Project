## HANDLE OUTLIERS

```python
def remove_outliers(df):
    df_copy = df.copy()
    thresholds={
        'bpm_zscore':4.0,
        'energy_zscore': 4.0,
        'danceability_zscore':4.0,
        'song_listeners_zscore': 3.0,
        'song_virality_zscore':3.0,
        'song_age_zscore': 5.0,
        'duration_minutes_zscore':4.0,
        'spectral_centroid_zscore': 4.0,
        'zero_crossing_rate_zscore': 4.0,
        'engagement_ratio_zscore': 3.0
    }
    print(f'Number of rows in the Dataframe before outlier removal: {len(df_copy)}')
    for col, threshold in thresholds.items():
        if col in df_copy.columns:
            df_copy=df_copy[abs(df_copy[col])<=threshold]
    print(f'Number of rows in the DataFrame after outlier removal: {len(df_copy)}')
    return df_copy
    


```

### OBSERVATIONS:
1. After about 1988 the curve starts accelerating; by the 1990s “peak” window the count doubles and triples within a few years. That shaded band captures the beginning of mainstream digital cataloging (CDs, early digital archives) and hints that the dataset is collecting more of the “long tail” even before streaming.
2. The red band around the 2010s+ marks the exponential surge where we go from a few thousand releases a year to nearly ten thousand by 2023. This is too steep to be organic artist growth alone—it screams infrastructure change (streaming platforms), and it also means the sample mean is overwhelmingly weighted by these modern releases.
3. Because the count in 2023 dwarfs every prior year, any aggregate metric will be skewed toward the present.
